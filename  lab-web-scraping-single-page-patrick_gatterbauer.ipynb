You have find data on the internet about currently popular songs. Billboard maintains a weekly Top 100 of "hot" songs here: http://www.popvortex.com/music/charts/top-100-songs.php.

It's a good place to start! Scrape the current top 100 songs and their respective artists, and put the information into a pandas dataframe.

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
pd.set_option('display.max_rows', None)
```

```python
r = requests.get('http://www.popvortex.com/music/charts/top-100-songs.php')
r.status_code
```

```python
# target url
#url = 'http://www.popvortex.com/music/charts/top-100-songs.php'
```

```python
# making requests instance
#reqs = requests.get(url)
```

```python
# using the BeautifulSoup module
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# displaying the title
titles = []

#add title name to the empty list titles
for i in soup.find_all(class_='title'):
    titles.append(i.get_text())
    
titles
```

```python
#add title name to the empty list artists
artists = []

for m in soup.find_all('em', class_="artist"):
    artists.append(m.get_text())
    
artists
```

```python
#create a dataframe
df = pd.DataFrame({'artists': artists, 'titles': titles})

#call the dataframe
df
```
